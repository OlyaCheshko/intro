```python
fig, axes = pit.subplots(2,3) #то есть две строки и три столбца
fig.set_figwidth(15)
fig.set_figheight(5)

axes[0,1].scatter(np.arange(20),np.arange(20)+3*randn(20), color='r') # randn generates random numbers from a normal distribution
axes[1,2].bar(np.arange(10),randn(10),color='m') # use randn(10) for the heights of the bars
```

```python
#При работе с объектом Subplots каждый график мы можем задавать в том числе и через циклы
fig,axes = pit.subplots(2, 2, sharex=False, sharey=False) #параметры sharex sharey заданные тру или фалс определяют объединять оси или нет 
fig.set_figheight(5)
fig.set_figwidth(15)
for i in range (2): 
    for j in range (2) :
        axes [i, j].hist(randn(500), bins=50, color='y', alpha=0.5)
#Достаточно популярный способ задания пустого пространства вокруг графиков
pit.subplots_adjust(wspace=0.0, hspace=0.0)
```
