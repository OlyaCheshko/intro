```python
import pandas as pd
```

```python
raw_data={'subject_id':[1,2,3,4,5], 'first_name':['Alex','Amy','Allen','Alice','Ayoung'],'last_name':['Anderson','Ackerman','Ali','Aoni','Atiches']}
df_a=pd.DataFrame(raw_data, columns=['subject_id','first_name','last_name'])
# df_a.index=[0,1,2,3,4] они и так будут такими
df_a
```

```python
df_a
```

```python
raw_data={'subject_id':[4,5,6,7,8], 'first_name':['Billy','Brian','Bryce','Betty','Ayoung'],'last_name':['Bonder','Black','Balwner','Brice','Btisan']}
df_b=pd.DataFrame(raw_data, columns=['subject_id','first_name','last_name'])
df_b.index=[2,3,4,5,6] 
df_b
```

```python
raw_data = {'subject_id': [1, 2, 3, 4, 5, 7, 8, 9, 10, 11], 'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_n = pd.DataFrame(raw_data, columns = ['subject_id', 'test_id'])
df_n
```

```python
df_a
```

```python
#Метод CONCAT
df_new=pd.concat([df_a,df_b],axis=0) #Объединение по строкам
df_new
```

```python
df_new_=pd.concat([df_a,df_b],axis=1) #Объединение по столбцам
df_new_
```

```python
#Метод CONCAT, но только с помощью атрибута "inner" 
df_new_=pd.concat([df_a,df_b],axis=1,join='inner') #остались только те строки, которые присутствуют по индексам в обоих таблицах
df_new_
```

```python
#Метод APPEND
df_a.append(df_b)
```

```python
#Метод JOIN - основан на объединении таблиц через индексы (способ объединения указывается с помощью параметра how=['left','right','inner','outer'] )
#df_a.join(df_b, how ='left')
df_a.join(df_b, rsuffix='_right_table', how='left') #то есть берутся все строки из датафрейма_а и к ним присоединяются все строки из датафрейма_б с теми же "совпадающими" индексами
#с помощью rsuffix ко всем столбцам содержащим инфу из правой(\второй\в данном случае датафрейма_б) объединенной таблицы добавляется суффикс 
```

```python
# Метод MERGE - используется для объединения таблиц по любым колонкам с помощью методой left_on и right_on 
pd.merge(df_new, df_n, on='subject_id')
```

```python
#стоит отметить что колонка subject_id не дублируется в новой таблице
pd.merge(df_new, df_n,on='subject_id') #по умолчанию используется how=inner 
```

```python
pd.merge(df_new, df_n,left_on='subject_id', right_on='subject_id')
```

```python
pd.merge(df_a, df_b,on='subject_id', how='left') #то есть берутся все строчки из а и все строчки из б с такими же совпадающими индексами
```

```python
pd.merge(df_a, df_b,on='subject_id', how='right') #наоборот, берутся все строчки из б и все строчки из а с такими же совпадающими индексами
```

```python
pd.merge(df_a, df_b,right_index=True, left_index=True)
```
