```python
# чтобы посмотреть какие колонки занимают больше всего памяти
for dtype in ['float64', 'int64', 'object','category','bool' ]:
    selected_dtype = df.select_dtypes(include=[dtype])
    mean_usage_b=selected_dtype.memory_usage(deep=True).mean ( )
    mean_usage_kb = mean_usage_b / 1024
    print( "Average memory usage for {} columns: {:03.2f} KB".format(dtype, mean_usage_kb) )
```

```python
#видим что колонки object занимают половину памяти, занимаемой всем датасетом
#попробуем оптимизировать память, сначала оптимизировав числовые колонки
df['age']=df['age'].astype('float32')
df['fare']=df['fare'].astype('float32')
df.info()
```

```python
#тип данных категория - позволяет более эффективно работать с категориальными признаками
[(col,df[col].nunique()) for col in df.columns]
```

```python
df_with_cat=df.copy()
df_with_cat['parch']=df_with_cat['parch'].astype('category')
df_with_cat['sibsp']=df_with_cat['sibsp'].astype('category')
df_with_cat.info()

#соптимизировали память практически на 20кб
```

```python
%%timeit
df.groupby('parch')['age'].mean().to_frame()
#посмотрим на увеличение скорости кода
```

```python
%%timeit
df_with_cat.groupby('parch')['age'].mean().to_frame()
#посмотрим на увеличение скорости кода
```
